```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split

# ----------------------------
# 1. Données : surface (m²) → prix (€)
# ----------------------------
X = np.array([[30], [40], [50], [60], [80]])  # 2D array
y = np.array([100000, 130000, 160000, 180000, 230000])  # 1D array

# ----------------------------
# 2. Séparer train et test
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# ----------------------------
# 3. Entraînement du modèle
# ----------------------------
model = LinearRegression()
model.fit(X_train, y_train)

# ----------------------------
# 4. Coefficients appris
# ----------------------------
w = model.coef_[0]   # pente
b = model.intercept_ # ordonnée à l’origine
print(f"Équation de la droite : y = {w:.2f}x + {b:.2f}")

# ----------------------------
# 5. Prédiction sur test
# ----------------------------
y_pred = model.predict(X_test)

# ----------------------------
# 6. Évaluation des performances
# ----------------------------
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE : {mse:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"MAE : {mae:.2f}")
print(f"R² : {r2:.2f}")

# ----------------------------
# 7. Prédiction d’un nouvel appart
# ----------------------------
nouvelle_surface = np.array([[70]])
prix_pred = model.predict(nouvelle_surface)
print(f"Prix estimé pour 70 m² : {prix_pred[0]:,.0f} €")

# ----------------------------
# 8. Visualisation
# ----------------------------
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Données')
plt.plot(X, model.predict(X), color='red', label='Modèle')
plt.xlabel("Surface (m²)")
plt.ylabel("Prix (€)")
plt.title("Régression linéaire - Prix d’un appartement")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# ----------------------------
# 1. Données
# ----------------------------
# [Surface (m²), Nombre de chambres, Étage]
X = np.array([
    [30, 1, 0],
    [40, 2, 1],
    [50, 2, 2],
    [60, 3, 3],
    [80, 3, 4],
    [100, 4, 5]
])

# Prix en euros
y = np.array([100000, 130000, 160000, 180000, 230000, 300000])

# ----------------------------
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X, y, test_size=0.2, random_state=0)
# ----------------------------
# ----------------------------
# 2. Entraînement du modèle
# ----------------------------
model_2 = LinearRegression()
model_2.fit(X_train_2, y_train_2)
# ----------------------------
# 3. Coefficients appris
# ----------------------------
w_2 = model_2.coef_
b_2 = model_2.intercept_
print(f"equation du modèle : y= {w_2[0]:.2f}x1 + {w_2[1]:.2f}x2 + {w_2[2]:.2f}x3 + {b_2:.2f}")
# ----------------------------
# 4. Prédiction sur test
# ----------------------------
y_pred_2 = model_2.predict(X_test_2)
# ----------------------------
# 5. Évaluation des performances
# ----------------------------
mse_2 = mean_squared_error(y_test_2, y_pred_2)
rmse_2 = np.sqrt(mse_2)
r2_2 = r2_score(y_test_2, y_pred_2)
print(f"MSE : {mse_2:.2f}")
print(f"RMSE : {rmse_2:.2f}")
print(f"R2 : {r2_2:.2f}")
# ----------------------------
# 6. Prédiction d’un nouvel appart
# ----------------------------
nouvel_appart_2 = np.array([[70, 3, 2]])
prix_pred_2 = model_2.predict(nouvel_appart_2)
print(f"Prix estimé pour le nouvel appart : {prix_pred_2[0]:,.0f} €")
# ----------------------------
# 7. Visualisation
# ----------------------------
# Note : Visualisation 3D pour la régression multiple
from mpl_toolkits.mplot3d import Axes3D
```
